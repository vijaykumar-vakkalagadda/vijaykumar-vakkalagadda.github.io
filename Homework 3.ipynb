```python
buildings2 = pd.read_csv("https://uiuc-ischool-dataviz.github.io/is445_AOUAOG_fall2021/week02/data/building_inventory.csv",na_values = {'Square Footage': 0,'Year Acquired': 0,'Year Constructed': 0})
# reading the date and filling the empty values with Null
```

```python
buildings2.plot(x="Year Acquired", y ="Year Constructed", figsize = (20,4), kind='scatter')
# I really like this plot as the data is alligned with the linear regression
# I am trying to constrct a scatter plot using year  acquired and year constructed
```

```python
buildings2.head()
```

```python
sfs = buildings2.groupby("Bldg Status")["Square Footage"].sum() # total square footage of building status is calculated
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(sfs.index, sfs.values) # using the calculated values to plot
ax.set_yscale('log') # showing larger scale range
ax.set_xlabel('Bldg Status')
ax.set_ylabel('Total Square Footage')
plt.show()
# Firstly i am calculating the sum of square footage of the building based on the building status and storing it in a variable "sfs". Then i am using the index and values from the stored variable to construct a visualization.
```

```python
b = buildings2.groupby("Year Acquired")["Square Footage"].median() # claculating the median square footage per acquired year
```

```python
fig, ax = plt.subplots(figsize=(20,5))
ax.plot(b.index, b.values) # using the calculated values to plot
ax.set_xlabel("Year acquired") # setting the x label
ax.set_ylabel("Median Sqrft") #setting the y label
plt.show()
# Firstly i am calculating the median of square footage of the building based on the year acquired and storing it in a variable "b". Then i am using the index and values from the stored variable to construct a visualization.
```
